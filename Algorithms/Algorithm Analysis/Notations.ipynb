# Asympotic Notations
Aysmpotic notations are languages that allows us to analyze an algorithm's running time by identifying its behaviour as the input size for the algorithm increases
* Big-oh (O)
* Big-Omega (Ω)
* Theta (Θ)
* Litle-oh (o)
* Litle-Omega (ω)

# Big-oh (O)
Big-oh notation gives the tight upper bound of the functions. It measure the worst time complexity of an algorithm possibly takes to complete
This means, a fuction f(x) is Big-oh funtion g(x) if there exits the +ve constant c and n0 such that:
f(n)< =c * g(x)
where c>0 and n0>=1 for all n>=n0

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fx = np.array([0, 4, 6, 8, 10, 12,15,16,17])
fy=np.array([2, 6, 7, 10, 11, 14,15,18,20])

# gx=[0, 4, 6, 8, 10, 12]

gy=np.array([0, 4, 8,12 , 15, 17,23,30,35])

plt.plot(fx, fy, 'b')
plt.plot(fx,gy,"b")
plt.xlabel('Input ->')
plt.ylabel('Time ->')
plt.title('BIg O')
plt.show()
```
